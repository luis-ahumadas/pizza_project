```python
import pandas as pd

# Load original dataset
df = pd.read_csv("pizza_sales.csv")

# Tells pandas that days come first (dayfirst=True) and handle mixed delimiters (/ vs -)
df['order_date'] = pd.to_datetime(df['order_date'], format='mixed', dayfirst=True)

# Format into standard format (YYYY-MM-DD)
df['order_date'] = df['order_date'].dt.strftime('%Y-%m-%d')

# Save the cleanly formatted data into a new file
df.to_csv('pizza_sales_fixed.csv', index=False)
```
